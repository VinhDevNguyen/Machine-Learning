```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = np.array([147, 150, 153, 160])
print(X)
```

```python
y = np.array([49, 53, 51,54])
print(y)
```

```python
plt.scatter(X,y)
```

```python
x_new = np.c_[np.ones((4, 1)), X]
print(x_new)
```

```python
minimize_theta = np.linalg.inv(x_new.T.dot(x_new)).dot(x_new.T).dot(y)
minimize_theta
```

```python
x1 = np.array([[147],[160]])
print(x1)
```

```python
x1_new = np.c_[np.ones((2, 1)), x1]
print(x1_new)
```

```python
y_predict = x1_new.dot(minimize_theta)
print(y_predict)
```

```python
plt.plot(x1, y_predict, "r-")
plt.plot(X, y, "b.")
```
