# Bài Tập 1


```python
import numpy as np 
import matplotlib.pyplot as plt
```

# khai báo 2 thư viện numpy và matplotlib
# numpy dùng để tạo dataset
# matplotlib dùng để vẽ đồ thị để thấy được tương quan của 2 đại lượng x, y và vẽ đồ thị linear regression

# tạo 1 bộ dữ liệu random cho X

```python
X = 2 * np.random.rand(100,1)
print(X)
```

# tạo 1 bộ dữ liệu random cho y có số chiều giống X và cộng thêm 1 bộ nhiễu gaussian 

```python
y = 4 + 3 * X + np.random.randn(100,1)
print(y)
```

# Vẽ biểu đồ biểu thị mối quan hệ giữ X và y

```python
plt.scatter(X,y)
plt.show()
```

# Để thực hiện tính toán hệ số bằng công thức Normal Equation ta thêm 1 cột số 1 trước X vì dạng tổng quát của Linear Regression là 
# y = $\theta$0 + $\theta$1 * x với x0 = 1

```python
X_b = np.c_[np.ones((100,1)), X]
print(X_b)
```

# áp dụng công thức Normal Equation để tính giá trị  $\theta$  minimize
# $\hat{\theta}$ = ($X^T$ X)$^-$$^1$ $X^T$ y

```python
minimize_theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print(minimize_theta)
```

# Kiểm tra độ chính xác của đồ thị, dự đoán kết quả y dựa trên X và minimize theta

```python
y_predict = X_b.dot(minimize_theta)
print(y_predict)
```

# Vẽ đồ thị để xem độ tương quan giữa giá trị thực tế và hàm Linear Regression

```python
plt.plot(X, y_predict, "r-")
plt.plot(X, y, "b.")
plt.show()
```

# Bài Tập 2

# Tạo mảng chứa dữ liệu của X

```python
X = np.array([147, 150, 153, 160])
print(X)
```

# Tạo mảng chứa dữ liệu của y

```python
y = np.array([49, 53, 51, 54])
print(y)
```

# Vẽ biểu đồ xem mối liên hệ giữa X và y

```python
plt.scatter(X, y)
```

# Thêm 1 cột giá trị 1 vào trước X

```python
x_new = np.c_[np.ones((4, 1)), X]
print(x_new)
```

# Tìm giá trị minimize theta bằng công thức Normal Equation

```python
minimize_theta = np.linalg.inv(x_new.T.dot(x_new)).dot(x_new.T).dot(y)
print(minimize_theta)
```

# Dự đoán giá trị y dựa trên các tham số theta mới tìm được

```python
y_predict = x_new.dot(minimize_theta)
print(y_predict)
```

# vẽ đồ thị để xem độ chính xác của Linear Regression với giá trị thực tế

```python
plt.plot(X, y_predict, "r-")
plt.plot(X, y, "b.")
```

```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
```

# Tách dữ liệu thành 2 phần X và y

# Bài Tập 4

```python
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

# khai báo thư viện pairplot để vẽ biểu đồ pairplot giữa các dữ liệu
# thư viện mpl_toolkits.mplot3d để vẽ đồ thị 3d

```python
data = pd.ExcelFile('demo_data_mul.xls')
```

```python
df = pd.read_excel(data, 0, header = 0)
df
```

```python
df.corr(method='pearson')
```

# Vẽ đồ thị để xem mối quan hệ giữa các đại lượng so với kết quả Stock_index_price

```python
sns.pairplot(data = df, kind = 'scatter')
plt.show()
```

# Ta thấy các đại lượng Year và Month ko ảnh hưởng nhiều đối với đại lượng Stock Index Price
# 2 đại lượng còn lại là Interest_rate và Unemployment_Rate ảnh hưởng nhiều đối với Stock_Index_price
# đối với Interest_rate thì khi giá trị tăng thì có xu hướng Stock_Index_price tăng
# Unemployment_Rate giảm thì Stock_Index_price có xu hướng giảm
# ta đi tìm phương trình hồi quy với 2 đại lượng Interest_rate và Unemployment_Rate

```python
x_train = df[['Interest_Rate', 'Unemployment_Rate']]
x_train
```

```python
y_train = df[['Stock_Index_Price']]
y_train
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
print(lr.intercept_)
print(lr.coef_)
```

```python
y_predict = lr.predict(x_train)
y_predict
```

# Vẽ đồ thị scatter để xem độ chính xác của Linear Regression


```python
plt.scatter(x_train[['Interest_Rate']], y_predict, color = 'red')
plt.scatter(x_train[['Interest_Rate']], y_train, color = 'blue')
```

```python
plt.scatter(x_train[['Unemployment_Rate']], y_predict, color = 'red')
plt.scatter(x_train[['Unemployment_Rate']], y_train, color = 'blue')
```

# Vẽ biểu đồ 3D để xem tổng quát giữ đại lượng thực tế với miền dự đoán của Linear Regression

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x_grid = np.linspace(min(x_train['Interest_Rate']), max(x_train['Interest_Rate']), 20)
y_grid = np.linspace(min(x_train['Unemployment_Rate']), max(x_train['Unemployment_Rate']), 20)
x_grid, y_grid = np.meshgrid(x_grid, y_grid)
exog = pd.core.frame.DataFrame({'Interest_Rate' : x_grid.ravel(), 'Unemployment_Rate' : y_grid.ravel()})
out = lr.predict(exog)

ax.plot_surface(x_grid, y_grid, out.reshape(20, 20), rstride = 1, cstride = 1, alpha = 0.3)

ax.scatter(x_train['Interest_Rate'], x_train['Unemployment_Rate'], y_train, color = 'red', marker = 'o', alpha = 1)
ax.set_xlabel('Interest_Rate')
ax.set_ylabel('Unemployment_Rate')
ax.set_zlabel('Stock_Index_Price')

ax.view_init(20, 75)
plt.show()
```
